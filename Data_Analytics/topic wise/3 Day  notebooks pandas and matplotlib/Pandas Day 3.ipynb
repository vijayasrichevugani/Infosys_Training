### Exercise - groupby 

```python
sys = ['s1','s1','s1','s1',
        's2','s2','s2','s2']
net_day = ['d1','d1','d2','d2',
        'd1','d1','d2','d2']
spd = [1.3, 11.4, 5.6, 12.3, 
        6.2, 1.1, 20.0, 8.8]
df = pd.DataFrame({'set_name':sys,
                    'spd_per_day':net_day,
                    'speed':spd})
```

```python
df["speed"]
```

```python
median_df = df.groupby(by=["set_name","spd_per_day"]).median()
```

```python
median_df
```

```python
median_df.columns
```

```python
median_df = median_df.rename(columns={"speed":"Median"})
```

## Pivot method - non duplicate indices
## Pivot table method - duplicate indices allowed but used with aggfunc

```python
df.pivot_table(columns="spd_per_day",values="speed",index="set_name",aggfunc="sum")
```
