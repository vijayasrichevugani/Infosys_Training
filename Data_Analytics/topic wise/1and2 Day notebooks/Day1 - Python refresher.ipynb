```python
def odd_num(n):
    l1=[]
    for i in range(1,n,2):
        l1.append(i)
    return l1
```

```python
print(odd_num(10))
```

```python
list(range(1,10,2))
```

```python
def reverse_text(text):
    return text[::-1]
print(reverse_text("Infosys"))
```

```python
def letter_count(text):
    dict1={}
    for i in text:
        dict1[i] = text.count(i)
    return dict1
```

```python
print(letter_count("infosys"))
```

#### List comprehension

```python
def odd_num_gen(n):
    return [i for i in range(0,n+1) if i%2==1]
```

```python
odd_num_gen(11)
```
